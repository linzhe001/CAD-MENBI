```python
import os

def rename_directories(base_path):
    for dir_name in os.listdir(base_path):
        if dir_name.startswith("NORMAL_2021L_"):
            new_name = dir_name.replace("NORMAL_2021L_", "NORMAL_2021_")
            os.rename(os.path.join(base_path, dir_name), os.path.join(base_path, new_name))
            print(f"Renamed '{dir_name}' to '{new_name}'")

# 设置您的目录路径
base_path = 'data/NORMAL/2021'

# 调用函数
rename_directories(base_path)
```

```python
import os

def rename_folders(base_path):
    for folder in os.listdir(base_path):
        if folder.startswith("NORMAL_2020L_"):
            old_path = os.path.join(base_path, folder)
            number_part = int(folder.split("_")[2])  # 提取数字部分
            new_folder_name = f"NORMAL_2020_{number_part + 28}"  # 计算新的数字部分
            new_path = os.path.join(base_path, new_folder_name)
            os.rename(old_path, new_path)
            print(f"Renamed {old_path} to {new_path}")

# 设置基础路径
base_path = "data/NORMAL/2020L"

# 调用函数
rename_folders(base_path)
```

```python
import pandas as pd

def rename_names_in_csv(file_path):
    # 读取CSV文件
    df = pd.read_csv(file_path)

    # 检查并更新“Name”列
    def rename_name(name):
        if name.startswith("NORMAL_2020L_"):
            number_part = int(name.split("_")[2])
            return f"NORMAL_2020_{number_part + 28}"
        return name

    df['Name'] = df['Name'].apply(rename_name)

    # 将更改后的数据写回文件
    df.to_csv(file_path, index=False)
    print("CSV file updated successfully.")

# 调用函数
file_path = 'data.csv'
rename_names_in_csv(file_path)
```

```python
import os
import cv2
from PIL import Image

# 定义输入和输出文件夹路径
input_folder = 'data'
output_folder = 'data_processed'

# 遍历患者种类
for category in os.listdir(input_folder):
    category_path = os.path.join(input_folder, category)
    
    # 遍历入院年份
    for year in os.listdir(category_path):
        year_path = os.path.join(category_path, year)
        output_year_path = os.path.join(output_folder, category, year)
        
        # 创建输出年份文件夹
        os.makedirs(output_year_path, exist_ok=True)
        
        # 遍历患者
        for patient in os.listdir(year_path):
            patient_path = os.path.join(year_path, patient)
            output_patient_path = os.path.join(output_year_path, patient)
            
            # 创建输出患者文件夹
            os.makedirs(output_patient_path, exist_ok=True)
            
            # 遍历图像文件
            i=0
            for image_file in os.listdir(patient_path):
                image_path = os.path.join(patient_path, image_file)
                
                original_image = cv2.imread(image_path)  # 读取文件
                image = original_image.copy()
                # 将图像转换为灰度并高斯去噪
                gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                blurred = cv2.GaussianBlur(gray, (3, 3), 0)

                # Canny边缘检测
                canny = cv2.Canny(blurred, 1, 30, 3, L2gradient=True) 
                # 使用闭运算连接中断的边缘,迭代运算三次
                canny = cv2.morphologyEx(canny, cv2.MORPH_CLOSE, kernel=(5, 5), iterations=3)

                # 查找图像轮廓
                cnts = cv2.findContours(canny.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
                cnts = cnts[0] if len(cnts) == 2 else cnts[1]

                # 查找每个轮廓的区域
                contour_sizes = [(cv2.contourArea(contour), contour) for contour in cnts]

                # 过滤出最大轮廓区域并裁剪出ROI区域
                if len(contour_sizes) > 0:
                    largest_contour = max(contour_sizes, key=lambda x: x[0])[1]
                    x, y, w, h = cv2.boundingRect(largest_contour)
                    cv2.rectangle(image, (x, y), (x + w, y + h), (36, 255, 12), 2)
                    ROI = original_image[y:y + h, x:x + w]

                # 计算图像名称
                i = i + 1
                image_name = str(i) + '.jpg'
                cv2.imwrite(os.path.join(output_patient_path, image_name), ROI)
                print(image_path, "-->", image_name)
```
